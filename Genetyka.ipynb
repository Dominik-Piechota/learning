```python
def base_DNA_cleaning(sequence):
    """Function that cleans the sequence from '-|N' and white spaces. Moreover, function makes sure 
    that the returned sequence is in capital letters.
    
    :param seq: str - sequence of DNA or RNA.
    :return: str - cleaned DNA or RNA.
    """

    sequence = sequence.upper()
    if sequence in ['-', 'N', ' ']:
        return ''
    elif sequence == 'U':
        return 'U'
    elif sequence == 'A': 
        return 'A'

    elif sequence == 'T':
        return 'T'

    elif sequence == 'C':
        return 'C'

    elif sequence == 'G':
        return 'G'
```

```python
def ATGC_count(sequence, rounding=None):
    """Function that counts the amount of certain nitrogenous principles and GC %, if u added rounding parameter, for default its None.
    
    :param sequence: str - sequence of DNA.
    :param rounding: int - determines the approximation. Default specified as None.
    :return: dictionary - amount of nitrogenous principles.
    """
    dictionary = dict(Counter(sequence))
    if rounding != None:
        perc = ((dictionary.get('G') + dictionary.get('C')) * 100) / sum(dictionary.values())
        perc = round(perc, int(rounding))
        dictionary.update({'%GC': perc})
    return dictionary
    
```

```python
def to_rna_element(sequence, seq=None):
    """Function that mainly transtripts DNA strand into RNA strand. There is one more function.
    If u add "DNA" as seq parameter, it will return DNA complementary to given one, not RNA.

    :param sequence: str - string containing the DNA strand.
    :param seq: str - its additional param. Can be None or DNA, others won't do anything special.
    :return: str - transcripted RNA/DNA string.
    """
    if (sequence == 'A') and (seq == 'DNA'):
        return 'T'

    elif (sequence == 'A') and (seq != 'DNA'): 
        return 'U'

    elif sequence == 'T':
        return 'A'

    elif sequence == 'C':
        return 'G'

    elif sequence == 'G':
        return 'C'
    else:
        raise ValueError("Probably your DNA strand needs some cleaning. There is something different from 'A|T|G|C' ")
```

```python
def proteins(sequence, lenght=None):
    '''This function provide Amino acids sequence in return based on given RNA sequence. 
    The translation is being stopped, when it hits one of STOP codons. The sequence does not need to be started
    with "AUG".
    
    :param strand: str - RNA sequence.
    :param lenght: str - 'short' or 'long', it determines whatever the output amino-acids are in long, full name like 
    "Methionine" or in short one like "M".
    :return: str - the sequence of amino-acids.
    '''
    
    codons = None
    separator = ''
    proteins = []
    if 'T' in sequence:
        raise ValueError('You provided DNA sequence instead of RNA sequence')
        
    if lenght == 'long':
        codons = codons_full
        separator = ', '
    elif (lenght == 'short') or (lenght == None):
        codons = codons_short  
    else:
        raise ValueError('wrong input in place of lenght. It has to be "long" for long name of amino-acids or "short" for the short ones.')
        
    for i in range(0, len(sequence), 3):
        protein = codons[sequence[i:i+3]]
        if protein == 'STOP':
            break
        proteins.append(protein)
        result = separator.join(proteins)
    return result
```

```python
def find_pattern(sequence, pattern):
    """Function that check for the specific pattern in the given sequence.

    :param sequence: str - sequence of DNA or RNA or other sequence.
    :param pattern: str - pattern of the shorter sequence like 'GAATTC'.
    :return: str - if there is occur of the pattern, it will return the beggining 
    index and final index of the pattern's nucleotides
    """
    import re
    import sys
    try:
        re.compile(pattern)
    except:
        print('Invalid pattern. Please make sure to input correct one')
        sys.exit()
        
    match = re.search(pattern, sequence)
    
    if match:
        print('Found the occur to the pattern   :', match.group())
        print('Starting at index :', match.start())
        print('Ending at index   :', match.end())
    else:
        print('Did not find the pattern.')
```

```python
def sequencing(sequence, pattern):
    """Function that check for the specific repetition in the given sequence.

    :param sequence: str - sequence of DNA or RNA or other sequence.
    :param pattern: str - pattern of the short sequence like 'GAATTC'.
    :return: str - if there is at least one occur of the pattern, it will return the part of sequence from the start to the hit of pattern.
    """
    
    amount = re.findall(pattern, sequence)
    occur = f'There is/are {len(amount)} occurencies of {pattern} found in. First on {sequence.index(pattern)}'
    if len(amount) >= 1:
        result = re.split(pattern, sequence)
        for number, match in enumerate(result[:-1]):
            print(f' Wzór sekwencji {number + 1}: {match + pattern}')
            print()
    return occur
```
